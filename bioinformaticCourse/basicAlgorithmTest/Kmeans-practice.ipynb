```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
from sklearn import datasets
```

```python
iris = datasets.load_iris()
data = iris.data
```

```python
data
```

```python
plt.scatter(data[:,0],data[:,1])
```

```python
def model(cluster,data):
    k_mean = KMeans(cluster)
    k_mean.fit(data)
    pre_label = k_mean.labels_
    return data, pre_label
```

```python
data_train , label = model(3,data)
```

```python
data_train0 = data_train[label==0]
data_train1 = data_train[label==1]
data_train2 = data_train[label==2]
plt.scatter(data_train0[:,0],data_train0[:,1],c = "red")
plt.scatter(data_train1[:,0],data_train1[:,1],c = "yellow")
plt.scatter(data_train2[:,0],data_train2[:,1],c = "blue")
```
